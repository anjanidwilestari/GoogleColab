```python
graph = {
  'A' : ['B','C'],
  'B' : ['C'],
  'C' : ['F'],
  'D' : [],
  'E' : ['F'],
  'F' : ['H'],
  'G' : ['H'],
  'H' : ['G']
}

visited = set()

def dfs(visited, graph, node): #function for BFS
  if node not in visited:
    print(node)
    visited.add(node)
    for neighbour in graph[node]:
      dfs(visited, graph, neighbour)

print("Hasil penelusuran graf menggunakan Depth-First Search")
dfs(visited, graph, 'A')        # function calling
```
