```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score #import libraby Cccuracy
from sklearn.metrics import matthews_corrcoef #import libraby Matthews Correlation Coefficient
from sklearn.linear_model import LogisticRegression #import Logistic Regression
```

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)

#Load data
mu = pd.read_csv('/content/drive/My Drive/Colab Notebooks/mushrooms.csv')

mu.head()
```

```python
#Encode label
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in mu.columns:
  mu[column]=le.fit_transform(mu[column])

mu.head()
```

```python
#SLice dataframe
X = mu.drop('class', axis='columns')
y = mu['class']

#Cek jumlah fitur dan instance
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lg = LogisticRegression(solver='lbfgs', max_iter=1000)

# Sesuaikan dt ke set training
lg.fit(X_train, y_train)

# Memprediksi label set test
y_pred_lg = lg.predict(X_test)

# Menghitung set accuracy
acc_lg = accuracy_score(y_test, y_pred_lg)

# Menghitung set Matthews correlation coefficient
mcc_lg = matthews_corrcoef(y_test, y_pred_lg)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_lg))
print(f"Test set accuracy: {acc_lg}")

print(" ")

#Print hasil Matthews correlation coefficient
print("Test set mcc: {:.2f}".format(mcc_lg))
print(f"Test set mcc: {mcc_lg}")
```
