```python
# Visualisasi Data 

data_mean=mean1, mean2, mean3, mean4, mean5, mean6
semester="1", "2", "3", "4", "5", "6"

print(data_mean)
print(semester)
```

```python
#Menampilkan data dengan Grafik Perbandingan Horizontal
plt.barh(semester, data_mean)
plt.xlabel('Nilai Rata-Rata')
plt.ylabel('Semester')
plt.title('Nilai Rata-Rata Setiap Semester Rapor Anjani')
plt.show()
```

```python
#Menampilkan data dengan Grafik Perbandingan Vertical

fig, ax = plt.subplots()

ax.bar(semester, data_mean, align='center')

ax.set_title('Nilai Rata-Rata Setiap Semester Rapor Anjani')
ax.set_ylabel('Nilai Rata-Rata')
ax.set_xlabel('Semester')

plt.show()

#Keterangan
#fig,ax â†’ Menambahkan kolom figure dan axes pada sublplots()
#ax.bar â†’ Membuat output persegi
#ax.set_title â†’ Judul grafik
#ax.set_ylabel â†’ Menamai garis vertikal
#ax.set_xlabel â†’ Menamai garis horizontal
#plt.show â†’ Menampilkan chart
```

```python
#Menampilkan data dengan Grafik Garis

plt.plot(semester, data_mean)

plt.xlabel('Semester')
plt.ylabel('Nilai Rata-Rata')
plt.title('Nilai Rata-Rata Setiap Semester Rapor Anjani')

plt.grid(True)

plt.show()

#Keterangan:
#plt.plot(data_mean, semester) â†’ Meletakkan posisi titik di x dan y
#plt.xlabel â†’ Keterangan sumbu horizontal
#plt.ylabel â†’ Keterangan sumbu vertikal
#plt.title â†’ Judul chart
#plt.grid â†’ Menampilkan garis grid
#plt.show â†’ Menampilkan chart
```

```python
#Menampilkan data dengan Diagram Kartesius

plt.title('Nilai Rata-Rata Setiap Semester Rapor Anjani')
plt.scatter(semester, data_mean,color="darkblue", marker='x', label='Nilai Mean Tiap Semester')


plt.xlabel('Semester')
plt.ylabel('Nilai Rata-Rata')

plt.grid(True)
plt.legend()

plt.show()

#Keterangan:
#plt.title â†’ Judul dari chart
#plt.scatter â†’ Properti pembentuk chart
#plt.xlabel â†’ Keterangan sumbu horizontal
#plt.ylabel â†’ Keterangan sumbu vertikal
#plt.grid â†’ Apabila bernilai True, maka akan ditampilkan garis grid
#plt.legend â†’ Titik-titik yang merupakan gabungan x dan y
#plt.show â†’ Menampilkan chart
```
