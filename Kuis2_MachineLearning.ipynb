```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)

#Load data
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/dataset_quiz.csv')

pd.set_option('display.max_columns', None)
df.head()
```

```python
#Pengecekan nilai #N/A
df.info()
```

```python
#Mengatasi data dengan missing values

#Membuang kolom id_anggota5 dikarenakan hanya ada 1 data yang non-null
del df['id_anggota5']

#Replace nilai missing values.
#Dikarenakan id tidak bisa menghasilkan nilai mean maka dari itu akan digantikan dengan nilai modus

#Kolom id_ketua
mode= df[df["id_ketua"]!="NaN"]["id_ketua"].mode()[0]
df['id_ketua'] = df['id_ketua'].replace("NaN", mode)

#Kolom id_anggota1
mode= df[df["id_anggota1"]!="NaN"]["id_anggota1"].mode()[0]
df['id_anggota1'] = df['id_anggota1'].replace("NaN", mode)

#Kolom id_anggota2
mode= df[df["id_anggota2"]!="NaN"]["id_anggota2"].mode()[0]
df['id_anggota2'] = df['id_anggota2'].replace("NaN", mode)

#Kolom id_anggota3
mode= df[df["id_anggota3"]!="NaN"]["id_anggota3"].mode()[0]
df['id_anggota3'] = df['id_anggota3'].replace("NaN", mode)

#Kolom id_anggota4
mode= df[df["id_anggota4"]!="NaN"]["id_anggota4"].mode()[0]
df['id_anggota4'] = df['id_anggota4'].replace("NaN", mode)


df.head()
```

Encode label

```python
#Encode label
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in df.columns:
  df[column]=le.fit_transform(df[column])

df.head()
```

**Seleksi Fitur**

```python
#Slice Dataframe
x = df.drop('keterlibatan_vltr', axis='columns')
y = df['keterlibatan_vltr']

#Cek jumlah fitur dan instance
x.shape
```

**Pemilahan data training, testing dan validasi**

```python
#Splitting Data
from sklearn.model_selection import train_test_split

#Data main akan dipisahkan menjadi data train dan data validation 
x_main,x_test,y_main,y_test=train_test_split(x,y,test_size=0.1,stratify=y)

x_train,x_val,y_train,y_val=train_test_split(x_main,y_main,test_size=0.1,stratify=y_main)
```

**2 Jenis Metode Klasifikasi Dengan Evaluasi Accuracy, Confusion Matrix, dan ROC**

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_curve

#inisialisasi model
svc = SVC(kernel='linear')

# Sesuaikan dt ke set training
svc.fit(x_train, y_train)

# Memprediksi label set test
y_pred_svc = svc.predict(x_val)

# Menghitung set accuracy
acc_svc = accuracy_score(y_val, y_pred_svc)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_svc))
print(f"Test set accuracy: {acc_svc}")

#Menghitung set Confusion Matrix
cm_svc = confusion_matrix(y_val, y_pred_svc)
cm_svc_display = ConfusionMatrixDisplay(confusion_matrix  = cm_svc, display_labels = [False, True])

#Print hasil Confusion Matrix
print("Confusion Matrix")
cm_svc_display.plot()
plt.show()

#Mendefinisikan set ROC
fpr, tpr, _ = roc_curve(y_val, y_pred_svc)

#Print hasil ROC curve
print("ROC Curve")
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

#visualisasi
sns.pairplot(df, hue="keterlibatan_vltr",palette="bright")
```

**Hyperparameter Tunning**

```python
#Implementasi Grid Search

from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {
    "C" : np.logspace(-1, 1, 3),
    "degree" : np.logspace(-1, 1, 3),
    "kernel" : ['rbf', 'poly']
    } 
  
grid = GridSearchCV(svc, param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
%timeit grid.fit(x_train, y_train)

#prediksi
y_pred_grid = grid.predict(x_val)

#evaluasi dengan score
score_grid = accuracy_score(y_val, y_pred_grid)

print(f'Hasil evaluasi: {score_grid}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

```python
#Implementasi Randomized Search

from sklearn.model_selection import RandomizedSearchCV
  
# defining parameter range
param_randomized = {
    "C" : np.logspace(-1, 1, 3),
    "degree" : np.logspace(-1, 1, 3),
    "kernel" : ['rbf', 'poly']
    } 
  
randomized = RandomizedSearchCV(svc, param_randomized, refit = True, verbose = 3)
  
# fitting the model for randomized search
%timeit randomized.fit(x_train, y_train)

#prediksi
y_pred_randomized = randomized.predict(x_val)

#evaluasi dengan score
score_randomized = accuracy_score(y_val, y_pred_randomized)

print(f'Hasil evaluasi: {score_randomized}')
print(f'Konfigurasi hyperparameter: {randomized.best_params_}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_curve

dt = DecisionTreeClassifier()

# Sesuaikan dt ke set training
dt.fit(x_train, y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(x_val)

# Menghitung set accuracy
acc_dt = accuracy_score(y_val, y_pred_dt)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")

#Menghitung set Confusion Matrix
cm_dt = confusion_matrix(y_val, y_pred_dt)
cm_dt_display = ConfusionMatrixDisplay(confusion_matrix  = cm_dt, display_labels = [False, True])

#Print hasil Confusion Matrix
print("Confusion Matrix")
cm_dt_display.plot()
plt.show()

#Mendefinisikan set ROC
fpr_dt, tpr_dt, _dt = roc_curve(y_val, y_pred_dt)

#Print hasil ROC curve
print("ROC Curve")
plt.plot(fpr_dt,tpr_dt)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

#visualisasi
plot_tree(dt)
```

**Hyperparameter Tunning**

```python
#Implementasi grid Search

from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grids = {
    'criterion':['gini','entropy'],
    'max_depth': list(range(5,11)),
    'max_leaf_nodes': list(range(5,11))
    } 
  
grids = GridSearchCV(dt, param_grids, refit = True, verbose = 3)
  
# fitting the model for grid search
%timeit grids.fit(x_train, y_train)

#prediksi
y_pred_grids = grids.predict(x_val)

#evaluasi dengan score
score_grids = accuracy_score(y_val, y_pred_grids)

print(f'Hasil evaluasi: {score_grids}')
print(f'Konfigurasi hyperparameter: {grids.best_params_}')
```

```python
#Implementasi Randomized Search

from sklearn.model_selection import RandomizedSearchCV
  
# defining parameter range
param_random = {
    'criterion':['gini','entropy'],
    'max_depth': list(range(5,11)),
    'max_leaf_nodes': list(range(5,11))
    } 
  
random = RandomizedSearchCV(dt, param_random, refit = True, verbose = 3)
  
# fitting the model for randomized search
%timeit random.fit(x_train, y_train)

#prediksi
y_pred_random = random.predict(x_val)

#evaluasi dengan score
score_random = accuracy_score(y_val, y_pred_random)

print(f'Hasil evaluasi: {score_random}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```
