Pada folder data, terdapat dataset jamur yang kita gunakan pada materi Decision Tree. Berdasarkan dataset yang sama, bandingkan performa antara algoritma DT dan RandomForest. Gunakan tunning hyperparameter untuk mendapatkan parameter dan akurasi yang terbaik

Import Library

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #import DT
from sklearn.ensemble import RandomForestClassifier #import RandomForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

Persiapan Data

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)

#Load data
mu = pd.read_csv('/content/drive/My Drive/Colab Notebooks/mushrooms.csv')

mu.head()
```

Encode Label

```python
#Encode label
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in mu.columns:
  mu[column]=le.fit_transform(mu[column])

mu.head()
```

Seleksi fitur

```python
#SLice dataframe
X = mu.drop('class', axis='columns')
y = mu['class']

#Cek jumlah fitur dan instance
X.shape
```

Split data training dan testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Training Decision Tree

```python
# Hyperparameter tuning in Decision Trees

#Pada kasus kali ini kita akan menggunakan parameter Max Depth
md = DecisionTreeClassifier(max_depth=30, random_state=42)

# Sesuaikan dt ke set training
md.fit(X_train, y_train)

# Memprediksi label set test
y_pred_md = md.predict(X_test)

# Menghitung set accuracy
acc_md = accuracy_score(y_test, y_pred_md)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_md))
print(f"Test set accuracy: {acc_md}")
print(f"depth (maximum) MD : {md.tree_.max_depth}") #This is a simple python code to find the depth (maximum) of a given binary tree
```

```python
# Hyperparameter tuning in Decision Trees

#Pada kasus kali ini kita akan menggunakan parameter Max leaf nodes
mf = DecisionTreeClassifier(max_leaf_nodes=30, random_state=42)

# Sesuaikan dt ke set training
mf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_mf = mf.predict(X_test)

# Menghitung set accuracy
acc_mf = accuracy_score(y_test, y_pred_mf)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_mf))
print(f"Test set accuracy: {acc_mf}")
print(f"depth (maximum) MF : {md.tree_.max_depth}") #This is a simple python code to find the depth (maximum) of a given binary tree
```

```python
# Hyperparameter tuning in Random Forest

#Pada kasus kali ini kita akan menggunakan parameter max_features
mfe = RandomForestClassifier(random_state=42, n_jobs=-1, max_features=None)

# Sesuaikan dt ke set training
mfe.fit (X_train, y_train)

# Memprediksi label set test
y_pred_mfe = mfe.predict(X_test)

# Menghitung set accuracy
acc_mfe = accuracy_score(y_test, y_pred_mfe)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_mfe))
print(f"Test set accuracy: {acc_mfe}")
```

```python
# Hyperparameter tuning in Random Forest

#Pada kasus kali ini kita akan menggunakan parameter min_samples_split
mss = RandomForestClassifier(random_state=42, n_jobs=-1, min_samples_split=20)

# Sesuaikan dt ke set training
mss.fit (X_train, y_train)

# Memprediksi label set test
y_pred_mss = mss.predict(X_test)

# Menghitung set accuracy
acc_mss = accuracy_score(y_test, y_pred_mss)

#Print hasil accuracy
print("Test set accuracy: {:.2f}".format(acc_mss))
print(f"Test set accuracy: {acc_mss}")
```
